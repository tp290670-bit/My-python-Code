```python
string = input("enter a string :")
new_str =""
for i in string:
  if i >='a' and i<='z':
    new_str +=chr(ord(i)-32)
  elif i>='A' and i<='z':
    new_str +=chr(ord(i)+32)
  else:
      new_str +=i
print(f"changed string:{new_str}")
```
